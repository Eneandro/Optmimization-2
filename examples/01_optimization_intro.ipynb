# 1 Unconstrained optimization

## Minimum of a Quadratic Function (One Variable)

#### Find a solution to the problem

$$
\min_{x} f(x)
$$

$$
\text{s.t. } x \in \mathbb{R}
$$

where $f$ is a differentiable function  $f : \mathbb{R} \rightarrow \mathbb{R}$.


### Example 1.1

$$
f(x) = ax^2 + bx + c
$$

Taking the derivative:

$$
f'(x) = 2ax + b
$$

The optimum occurs when:

$$
2ax + b = 0 \;\Rightarrow\; x^* = -\frac{b}{2a}
$$


```python
a_, b_, c_ = 1.0, 0.0, 0.0

def plot_quadratic(a, b, c):
    global a_, b_, c_
    a_, b_, c_ = a, b, c
    x = np.linspace(-10, 10, 400)
    y = a*x**2 + b*x + c
    
    x_opt = -b/(2*a)
    y_opt = a*x_opt**2 + b*x_opt + c
    
    plt.figure()
    plt.plot(x, y)
    plt.scatter(x_opt, y_opt)
    plt.axvline(x_opt, color='red', linestyle='--')
    plt.axhline(y_opt, color='red', linestyle='--')

    plt.xlabel("x")                      # X axis label
    plt.ylabel("f(x)")                   # Y axis label
    plt.axhline(0, linewidth=1, color = 'black')
    plt.axvline(0, linewidth=1, color = 'black')

    plt.title(f"f(x) = ax² + bx + c \n f(x) = {a:.2f}x² + {b:.2f}x + {c:.2f}  \n Minimum at x = {x_opt:.2f} with f(x) = {y_opt:.2f}")
    plt.grid(True)
    plt.show()

widgets.interact(plot_quadratic,
                 a=(0.4,5.0,0.5),
                #  a=(-5.0,-0.4,0.5),
                 b=(-10.0,10.0,0.5),
                 c=(-50.0,50.0,1.0))

print("---------------------")
```

![Sym](../images/Sympy.jpg)

[SymPy](https://www.sympy.org/en/index.html) is a Python library for symbolic mathematics. It aims to become a full-featured computer algebra system (CAS) while keeping the code as simple as possible in order to be comprehensible and easily extensible. SymPy is written entirely in Python.

```python
x = 10
print(f" {x} {type(x)}")
```

```python
import sympy as sp
print(sp.__version__)

# Define symbolic variable
x = sp.symbols('x')
print(f" {x} {type(x)}")

# Defining the quadratic function
y = a_*x**2 + b_*x + c_

# Derivative and critical points
df = sp.diff(y, x)

# Solve for critical points
crit = sp.solve(df, x)

# Print the result of the critical points
print(f"Critical points: {crit}")

# Get the optimal value
x_opt = crit[0]
y_opt = y.subs(x, x_opt)

# Print the results
print(f"Minimum at x = {x_opt:.2f} with f(x) = {y_opt:.2f}")
```

## 2. Optimization with a Linear Constraint (Substitution Method)

Maximize:

$$f(x,y) = xy \qquad \text{Subject to:} \qquad x + y = C$$

Substituting:

$$
y = C - x
$$

$$
f(x) = x(C - x)
$$

Taking the derivative:

$$
f'(x) = C - 2x
$$

Optimum:

$$
x^* = \frac{C}{2}, \quad y^* = \frac{C}{2}
$$

### 1️ Problem Statement

We want to optimize a function:

$$
\text{Maximize or Minimize } f(x,y)
$$

subject to a constraint:

$$
g(x,y) = 0
$$

This is called **constrained optimization**.

### 2 Geometric Interpretation (Core Insight)

At the constrained optimum:

$$
\nabla f(x,y) = \lambda \nabla g(x,y) \qquad \text{where the gradient of a function} \quad f \quad \text{is} \quad \nabla f(x,y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
$$

### Why?

- $ \nabla f $ points in the direction of maximum increase of  $f$.
- $ \nabla g $ is perpendicular to the constraint curve.
- On the constraint surface, movement is tangent to:

$$
g(x,y) = 0
$$

At the optimum, the gradient of $f$ cannot have a tangential component; otherwise, we could increase or decrease $f$ while remaining feasible.

```python
# Defining the symbols for Lagrange multipliers
x, y, lam = sp.symbols('x y lam')

#Expression for the function and the constraint
f = x*y
g = x + y - 10

# Lagrangian
L = f - lam*g

# Solving the system of equations
eqs = [
    sp.diff(L, x),
    sp.diff(L, y),
    sp.diff(L, lam)
]

#Get the solution
sol = sp.solve(eqs, [x, y, lam])
print(f"Solution = {sol}")
```
