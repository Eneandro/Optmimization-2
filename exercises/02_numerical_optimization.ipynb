### Function definition

```python
x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)

X,Y = np.meshgrid(x,y)

Z = 20-X**2+Y**2-40*(np.tan((2*np.pi*X/50)/ (2*np.pi*Y/50))) # Adapt here the function to be optimized
```

### Plot in 2D

```python
plt.figure(figsize=(6,5))
plt.contourf(X,Y,Z,50)
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Potential Surface")
plt.show()
```

### Plot in 3D

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(
    X, Y, Z,
    cmap='viridis',
    edgecolor='none',
    alpha=0.9
)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.set_title('Surface Plot')

fig.colorbar(surface, shrink=0.5, aspect=10)

plt.show()
```
